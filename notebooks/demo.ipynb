# Demo rapido (assets incluidos)

Este notebook genera un PDF demo usando `examples/demo.yaml` y los assets incluidos.

```python
from pathlib import Path
import yaml
from pdfgen_juanipis.api import render_with_defaults

root = Path('..').resolve()
data = yaml.safe_load((root / 'examples' / 'demo.yaml').read_text(encoding='utf-8'))
out_path = root / 'examples' / 'demo.pdf'
render_with_defaults(data, output_path=out_path, root_dir=root)
out_path
```

Si no se ve el PDF en tu entorno, abre manualmente `examples/demo.pdf`.
