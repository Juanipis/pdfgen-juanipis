# pdfgen avanzado

Ejemplo con tablas, referencias inline y salida en bytes.

```python
from pdfgen.api import render_with_defaults_bytes

data = {
  "refs_catalog": {
    "1": "Fuente A (2024)",
    "2": "Fuente B (2024)"
  },
  "theme": {
    "header_banner_path": "banner.png",
    "header_logo_path": "logo.png",
    "title_line1": "Reporte de Indicadores",
    "title_line2": "Ejemplo Avanzado",
    "footer_site": "example.org",
    "footer_phone": "Contacto: +1 555 0100"
  },
  "sections": [
    {
      "title": "I. Resumen",
      "content": [
        {"type": "text", "text": "Texto con refs [1] y [2]."},
        {"type": "table", "table": {
          "groups": [
            {"title": "Consumo", "months": ["Ene", "Feb", "Mar"]},
            {"title": "Estrategias", "months": ["Ene", "Feb", "Mar"]}
          ],
          "rows": [
            {"dep": "Bogota", "vals": ["120", "128", "132", "95", "97", "101"]},
            {"dep": "Antioquia", "vals": ["86", "90", "94", "70", "72", "75"]}
          ]
        }}
      ]
    }
  ]
}

pdf_bytes = render_with_defaults_bytes(data)
with open('salida_avanzada.pdf', 'wb') as f:
    f.write(pdf_bytes)
```
