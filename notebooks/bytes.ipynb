# pdfgen bytes

Ejemplo para obtener el PDF como bytes y guardarlo manualmente.

```python
from pdfgen.api import render_with_defaults_bytes

data = {
    "theme": {
        "header_banner_path": "banner.png",
        "header_logo_path": "logo.png",
        "title_line1": "Reporte de Indicadores",
        "title_line2": "Ejemplo Bytes",
        "footer_site": "example.org",
        "footer_phone": "Contacto: +1 555 0100"
    },
    "sections": [
        {
            "title": "I. Introduccion",
            "content": [
                {"type": "text", "text": "Texto con <strong>negrita</strong>."}
            ]
        }
    ]
}

pdf_bytes = render_with_defaults_bytes(data)
with open("salida_bytes.pdf", "wb") as f:
    f.write(pdf_bytes)
```
